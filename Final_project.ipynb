```python
import pandas as pd

def calculate_quarterly_returns(data):
    data['Dates'] = pd.to_datetime(data['Dates'])
    data.set_index('Dates', inplace=True)

    # Calculate weekly returns
    weekly_returns = data.pct_change()

    # Calculate quarterly returns
    quarterly_returns = weekly_returns.rolling(window=4).agg(lambda x: (1 + x).prod() - 1)

    return quarterly_returns

# Read the dataset from Excel file
df = pd.read_csv("industry.csv")

# Calculate Quarterly Returns
quarterly_returns_df = calculate_quarterly_returns(df)
display(quarterly_returns_df)

```
